```python
merged_df['brand_product_scent'] = merged_df['brand_product'] + ',' + merged_df['all_scent']
merged_df
```

```python
import pandas as pd

# Group by product_name and concatenate all reviews
consolidated_desc = merged_df.groupby('brand_name')['brand_product_scent'].apply(','.join).reset_index()
consolidated_desc
```

```python
# Convert to dictionary
brand_scent_dict = consolidated_desc.set_index('brand_name')['brand_product_scent'].to_dict()
```

#### should preprocess

### clustering?

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Preprocess & Vectorize
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(consolidated_desc['brand_product_scent'])

# Choose Number of Clusters
num_clusters = 3
# (e.g., using the Elbow Method or Silhouette Score)

# Apply K-Means Clustering
model = KMeans(n_clusters=num_clusters, init='k-means++', max_iter=100, n_init=1, random_state=89)
model.fit(X)

# Review Output
order_centroids = model.cluster_centers_.argsort()[:, ::-1]
terms = vectorizer.get_feature_names_out()
for i in range(num_clusters):
    print(f"Cluster {i}:")
    for ind in order_centroids[i, :10]:  # print top 10 words per cluster
        print(f' {terms[ind]}')
    print('\n')
```

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import seaborn as sns

# Get cluster assignments for each data point
cluster_assignments = model.labels_

# Reduce data to 2D and plot it
tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X)

sns.scatterplot(X_tsne[:, 0], X_tsne[:, 1], hue=cluster_assignments, palette="viridis", legend="full")
plt.title('t-SNE Cluster Visualization')
plt.show()
```

```python
# Getting the cluster labels for each document
cluster_labels = model.labels_

# Mapping cluster labels to documents
clustered_documents = {i: [] for i in range(model.n_clusters)}
for i, label in enumerate(cluster_labels):
    clustered_documents[label].append(consolidated_desc['brand_name'][i])

# Now, `clustered_documents` is a dictionary where the keys are cluster labels 
# and the values are lists of documents belonging to that cluster.
clustered_documents
```

```python
# Creating a reverse lookup dictionary to map brand names to cluster labels
brand_to_cluster = {}
for cluster, brands in clustered_documents.items():
    for brand in brands:
        brand_to_cluster[brand] = cluster

# Mapping the brand names in the DataFrame to the cluster labels and creating a new column
merged_df['cluster'] = merged_df['brand_name'].map(brand_to_cluster)

merged_df
```

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity


# Initialize a CountVectorizer
vectorizer = CountVectorizer()

# Fit and transform the text data
X = vectorizer.fit_transform(consolidated_desc['brand_product_scent'])

# Compute cosine similarity
similarity_matrix = cosine_similarity(X, X)

similarity_df_2 = pd.DataFrame(similarity_matrix, index=brand_scent_dict.keys(), columns=brand_scent_dict.keys())
similarity_df_2
```

```python
lower_bound = 0.2
upper_bound = 0.5

# Apply the filter
filtered_similarity_df = similarity_df_2.where((similarity_df_2 >= lower_bound) & (similarity_df_2 <= upper_bound))

# Or if you want to replace NaN values with some placeholder value (e.g., 0):
filtered_similarity_df_filled = similarity_df_2.where((similarity_df_2 >= lower_bound) & (similarity_df_2 <= upper_bound), other=0)

filtered_similarity_df_filled
```

```python
# Step 1: Aggregate Reviews by Brand and Cluster
aggregated_reviews = merged_df.groupby(['brand_name', 'cluster'])['reviews'].agg(' '.join).reset_index()

# Initialize an empty DataFrame to store the results
results_df = pd.DataFrame(columns=['brand1', 'brand2', 'similarity_score', 'cluster'])

# For each cluster
unique_clusters = merged_df['cluster'].unique()
for cluster in unique_clusters:
    cluster_data = aggregated_reviews[aggregated_reviews['cluster'] == cluster]
    
    # Step 2: Create Document Vectors
    vectorizer = CountVectorizer()
    X = vectorizer.fit_transform(cluster_data['reviews'])
    
    # Step 3: Calculate Similarity
    similarity_matrix = cosine_similarity(X, X)
    
    # Convert to DataFrame for better visualization
    similarity_df = pd.DataFrame(similarity_matrix, 
                                 index=cluster_data['brand_name'], 
                                 columns=cluster_data['brand_name'])
    
    # Melt similarity_df into a long format
    melted_similarity_df = similarity_df.reset_index().melt(id_vars='brand_name', value_name='similarity_score')
    
    # Rename columns and add cluster number
    melted_similarity_df.columns = ['brand1', 'brand2', 'similarity_score']
    melted_similarity_df['cluster'] = cluster
    
    # Append to the results DataFrame
    results_df = pd.concat([results_df, melted_similarity_df], ignore_index=True)

# Drop rows where brand1 == brand2
results_df = results_df.loc[results_df['brand1'] != results_df['brand2']]
```
